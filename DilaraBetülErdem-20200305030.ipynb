# One-Hot Encoding Implementation

This Python code is a simple example of how you can implement **One-Hot Encoding**. Using the **Pandas** library, the `get_dummies()` function is used to convert categorical data to numeric data.

## Steps:
1. **Data Creation:**
- A sample DataFrame is created with a categorical column named 'Type' and a numeric column named 'Rating'.

2. **One-Hot Encoding Implementation:**
- The `pd.get_dummies()` function is used to convert the categorical data in the 'Type' column to numeric data by creating a separate column for each unique value.
- The first column (feature) is dropped using the `drop_first=True` parameter, thus preventing the "dummy variable trap".
